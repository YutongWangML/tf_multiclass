## Overview

This notebook plots the losses in the binary case, in which case the losses have 1-dimensional domain. (In general, the domain of the loss function is $\mathbb{R}^{k-1}$ where $k$ is the number of classes.)

```python
import sys
import os
sys.path.insert(1, os.path.join(sys.path[0], '../..'))
```

```python
import tf_multiclass.losses as losses
import keras.backend as K
import tensorflow as tf
import numpy as np
from tf_multiclass.utils import to_t_categorical
import matplotlib.pyplot as plt
```

```python
y_pred = tf.expand_dims(tf.linspace(-3.0,3.0,100),axis=-1)
y_true = to_t_categorical(np.ones((100,1)), num_classes=2)
```

```python
ww = losses.WWHinge(reduction='none')(y_true,y_pred)
dkr = losses.DKRHinge(reduction='none')(y_true,y_pred)
ce = losses.CrossEntropy(reduction='none')(y_true,y_pred)
```

```python
plt.plot(y_pred.numpy(), ww.numpy(), "k:")
plt.plot(y_pred.numpy(), dkr.numpy(), "k--")
plt.plot(y_pred.numpy(), ce.numpy(), '-.k')
plt.legend(['WW Hinge', 'DKR Hinge', 'Logistic'])
plt.show()
```
